```python
#DATAFRAME JOINS


#DataFrame1

import pandas as pd
import numpy as np
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6,8]),'Product':pd.Series(['oven','oven','oven','Television','Television','Television','Mobile'])}
df1 = pd.DataFrame(d1)


#DataFrame2

d2 = {'Customer_id':pd.Series([2,4,6,7,8]),'City':pd.Series(['Coimbatore','coimbatore','Chennai','Salem','Ooty'])}
df2=pd.DataFrame(d2)

print(df1)
print("\n")
print(df2)
print("\n INNER_JOIN")

inner_join_df=pd.merge(df1,df2,on='Customer_id',how='inner')#innerjoin
print(inner_join_df)

print("\n OUTER_JOIN")

outer_join_df=pd.merge(df1,df2,on='Customer_id',how='outer')#outerjoin
print(outer_join_df)

print("\n RIGHT_JOIN")

right_join_df=pd.merge(df1,df2,on='Customer_id',how='right')#rightjoin
print(right_join_df)

print("\n LEFT_JOIN")

left_join_df=pd.merge(df1,df2,on='Customer_id',how='left')#leftjoin
print(left_join_df)
```

```python
left_join_df=pd.merge(df2,df1,on='Customer_id',how='left')#leftjoin
print(left_join_df)
```

```python
left_join_df=pd.merge(df1,df2,how='cross')#leftjoin
print(left_join_df)
```

```python
#Dataframe

data={'Name':['Dima','James','Mia','Emity','Roben','John','Jordan'],
      'Salary':[50000,45000,51000,60000,41000,21450,34000],
      'Gender':['F','M','F','F','M','M','M'],
      'Age':[23,34,36,29,28,25,30]}
df_emp=pd.DataFrame(data)
df_emp
```

```python
#reshape using pandas(Melt()function)

df_melt=df_emp.melt()
df_melt
```

```python
#reshape using pandas(Melt()function)

df_melt=df_emp.melt(id_vars=['Gender'])
df_melt
```

```python
#reshape using pandas(Melt()function)

df_melt=df_emp.melt(id_vars=['Gender','Age'])
df_melt
```

```python
#reshape using pandas(Melt()function)

df_melt=df_emp.melt(id_vars=['Gender','Age','Name'])
df_melt
```

```python
#reshape using pandas(Melt()function)

df_melt=df_emp.melt(id_vars=['Gender'],value_vars=['Age','Salary'])
df_melt
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.close("all")
ts=pd.Series(np.random.randn(365),index=pd.date_range("2/15/2022",periods=365))
print(ts)
ts=ts.cumsum()#cummulative sum
print(ts)
ts.plot();
```

```python
#continuation of above program

df=pd.DataFrame(np.random.randn(365,6),index=ts.index,columns=list("ABCDEF"))
df=df.cumsum()
print(df)
plt.figure();
df.plot();
```

```python
#continuation of above program

df2=pd.DataFrame(np.random.rand(10,4),columns=["a","b","c","d"])
print(df2)
df2.plot.bar();
df2.plot.bar(stacked=True);
df2.plot.barh(stacked=True);
df2.plot.bar(stacked=False);
```
