```python
#subplots

import matplotlib.pyplot as plt
import numpy as  np

fig,ax=plt.subplots(2,2)
x=np.linspace(0,8,1000)

ax[0,0].plot(x,x**2,'g') #row=0 column=0
ax[0,0].title.set_text("Axis [0,0]")

ax[1,0].plot(x,x**3,'k') #row=1,column=0
ax[1,0].title.set_text("Axis [1,0]")

ax[0,1].plot(x,np.sin(x),'b') #row=0,column=1
ax[0,1].title.set_text("Axis [0,1]")

ax[1,1].plot(x,np.cos(x),'r') #row=1,column=1
ax[1,1].title.set_text("Axis [1,1]")
```
