```python
import sklearn
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
iris
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.60,test_size=0.40,random_state=47)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
predicted_y=model.predict(xtest)
```

```python
predicted_y
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
accuracy_score(ytest,predicted_y)
```

```python
confusion_matrix(ytest,predicted_y)
```

```python
z = classification_report(ytest,predicted_y)
print(z)
```

```python
#improve accuracy using PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
```

```python
pca.fit(xtrain)
```

```python
xtrain_transformed=pca.transform(xtrain)
```

```python
print(xtrain_transformed.shape)
```

```python
xtrain_transformed 
```

```python
pca.fit(xtest)
```

```python
xtest_transformed=pca.transform(xtest)
```

```python
print(xtest_transformed.shape)
```

```python
xtest_transformed 
```

```python
new_model=LogisticRegression(penalty='l1',solver='liblinear')
```

```python
new_model.fit(xtrain_transformed,ytrain)
```

```python
new_prediction=new_model.predict(xtest_transformed)
```

```python
new_prediction
```

```python
ytest,predicted_y,new_prediction
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
new_accuracy=accuracy_score(new_prediction,ytest)
new_accuracy
```

```python
confusion_matrix(ytest,new_prediction)
```

```python
z = classification_report(ytest,new_prediction)
print(z)
```
